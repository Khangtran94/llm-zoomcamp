### Prepare dataset:

### Index using minsearch:

```python
## “Build the searchable index from my dataset using the rules I defined earlier.”
from minsearch import Index
index = Index(
    text_fields = ['question','section','answer'],
    keyword_fields = ['course'])

index.fit(documents)
```

```python
## keyword_fields = filters (hard rules)
## These are used to limit the dataset BEFORE searching => ✅ exact match only

## 🟡 text_fields = search content (ranking area)
## These are used for: “Find the most relevant documents based on similarity”

# 🔥 In RAG terms
# keyword_fields	select dataset subset
# text_fields	find best matching documents

## boost_dict: how important each field is when scoring matches
```

```python
question = "I just discovered the course. Can I join now?"

search_results = index.search(
    question,
    boost_dict={"question": 2.0, "section": 0.5},
    filter_dict={"course": "llm-zoomcamp"},
    num_results=3
)

search_results
```

```python
[doc['question'] for doc in search_results]
```

### 1. Build search function for RAG:

```python
### boost_dict to emphasize the information; num_result = 5 (default)
def search(question, course = 'llm-zoomcamp', num_results=5):
    boost_dict = {"question": 2.0, "section": 0.5}
    filter_dict = {"course": course}

    return index.search(
        question,
        boost_dict=boost_dict,
        filter_dict=filter_dict,
        num_results = num_results)
```

#### Instructions vs prompt template:

```python
INSTRUCTIONS = """
Your task is to answer questions from the course participants
based on the provided context.

Use the context to find relevant information and provide accurate
answers. If the answer is not found in the context,
respond with "I don't know."
"""

USER_PROMPT_TEMPLATE = """
Question:
{question}

Context:
{context}
"""
```

### Format the output form search_results

```python
def build_context(search_results):
    lines = []

    for doc in search_results:
        lines.append(doc["section"])
        lines.append("Q: " + doc["question"])
        lines.append("A: " + doc["answer"])
        lines.append("")

    return "\n".join(lines).strip()
```

### 2. Build build_prompt function for RAG:

```python
def build_prompt(question, search_results):
    #### make the format better
    context = build_context(search_results)

    #### add the prompt
    prompt = USER_PROMPT_TEMPLATE.format(
        question=question,
        context=context
    )
    return prompt.strip()
```

### Define LLM function:

```python
from google.genai import types

def llm(prompt, model = "gemini-2.5-flash"):
    response = client.models.generate_content(
        model=model,
        contents=prompt,
        config=types.GenerateContentConfig(
            system_instruction=INSTRUCTIONS
        )
    )
    return response.text
```

## Full RAG

```python
def rag(query):
    ### Search + index inside search
    search_results = search(query)
    ### Prepare format
    prompt = build_prompt(query, search_results)
    ### push the prompt to LLM
    answer = llm(prompt)
    return answer
```

## Search function for AI Agent:

```python
search_tool = {
    "type": "function",
    "name": "search",
    "description": "Search the FAQ database for entries matching the given query.",
    "parameters": {
        "type": "object",
        "properties": {
            "query": {
                "type": "string",
                "description": "Search query text to look up in the course FAQ."
            }
        },
        "required": ["query"],
        "additionalProperties": False
    }
}
```

add functions to llm 

```python
def llm(prompt, model="gemini-2.5-flash"):
    response = client.models.generate_content(
        model=model,
        contents=prompt,
        config=types.GenerateContentConfig(
            system_instruction=INSTRUCTIONS,
            tools=[search_tool]
        )
    )

    return response
```

```python
for part in response.candidates[0].content.parts:
    if part.function_call:
        args = dict(part.function_call.args)

        results = search(**args)

        print(results)
```
